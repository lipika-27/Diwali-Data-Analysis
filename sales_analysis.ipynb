# Data Analysis

```python
df.columns
```

```python
gender = sns.countplot(x='Gender',data = df)
```

```python
gender = sns.countplot(x='Gender',data = df)

for values in gender.containers:
    gender.bar_label(values)
```

```python
sales_gender=df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gender)
```

# Age

```python
age_group = sns.countplot(x='Age Group', data = df,hue='Gender')

for values  in age_group.containers:
    age_group.bar_label(values)
```

```python
sales_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_age
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
# total number of order from top 10 cities
sales_states = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sales_states
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x='State',y='Orders',data=sales_states)
```

*From the above graph we can see that most of the orders are from Uttar Pradesh, Maharashtra, Karnataka respectively*

# Marital_Status

```python
sns.set(rc={'figure.figsize':(5,5)})
sales_mrg = sns.countplot(x='Marital_Status',data=df)
```

```python
# number of orders on the basis of marital status
mrg_order = df.groupby(['Marital_Status'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
mrg_order
```

```python
sns.barplot(x='Marital_Status',y='Orders',data=mrg_order)
```

```python
mrg_order_gender = df.groupby(['Marital_Status','Gender'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
mrg_order_gender
sns.barplot(x='Marital_Status',y='Orders',data=mrg_order_gender,hue='Gender')
```

*From the above graph we can see that most of the order placed by the married women*

# Occupation

```python
sns.set(rc={'figure.figsize':(21,6)})
occ = sns.countplot(x='Occupation',data=df)
for values in occ.containers:
    occ.bar_label(values)
```

```python
sales_occ = df.groupby(['Occupation'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(x='Occupation',y='Orders',data=sales_occ)
```

*From above graph we can see that most of the orders placed by the peoples working in IT sector, Healthcare and Aviation sector*

# Product Catagory

```python
sns.set(rc={'figure.figsize':(25,5)})
prd_cat = sns.countplot(x='Product_Category',data=df)

for v in prd_cat.containers:
    prd_cat.bar_label(v)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
prd_cat_amt= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=prd_cat_amt)
```

*The most sold products are Food, Clothing and Electronics*

# Conclusion
*Married women are from age group 26-35 and from state UP, Maharastra and Karnataka working in IT, Healthcare and Aviation sector are more likely to place order Food, Clothing and Electronics items.*
